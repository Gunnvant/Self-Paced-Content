# Statistical Plots (Pandas)

- Boxplot
- Density
- Histogram

```python
import pandas as pd
weather = pd.read_csv("weather.csv")
weather['date'] = pd.to_datetime(weather['date'])
weather['months'] = weather['date'].dt.month
weather.head(2)
```

```python
atlanta = weather.query("city=='Atlanta'")[['TMIN','TMAX','months']]
```

## Boxplot

```python
atlanta['TMAX'].plot(kind='box')
```

```python
atlanta.boxplot(column="TMAX",by='months')
```

## Histogram

```python
atlanta['TMAX'].plot(kind = 'hist')
```

## Density

```python
atlanta['TMAX'].plot(kind="density")
```

## Plots using Seaborn

```python
import seaborn as sns
```

## Boxplot

```python
sns.boxplot(x = atlanta['TMAX'])
```

```python
sns.boxplot(data = atlanta ,x = 'months',y = 'TMAX')
```

## Histogram

```python
sns.histplot(x=atlanta['TMAX'])
```

```python
sns.displot(atlanta,x='TMAX',col='months')
```

## Density Plot

```python
sns.displot(data=atlanta,x='TMAX',kind='kde')
```

```python
sns.displot(data=atlanta,x='TMAX',kind='kde',col="months")
```
