## Visualizing trend over time

```python
import pandas as pd
data = pd.read_csv("flights.csv")
```

```python
data.head(2)
```

```python
data.query("month=='Jan'").plot(x='year',y='passengers',kind='line')
```

```python
import seaborn as sns
```

```python
sns.lineplot(data.query("month=='Jan'"), x="year", y="passengers")
```

```python
sns.lineplot(data, x="year", y="passengers",hue='month')
```
