## Modifying seaborn plots

- Custom x axis and y axis labels and titles
- Changing the size of the plot

```python
import pandas as pd
age_sal = pd.read_csv("age_sal.csv")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.relplot(data=age_sal, x="Age", y="Annual Salary")
```

```python
plt.figure(dpi=200)
sns.relplot(data=age_sal, x="Age", y="Annual Salary")
plt.title("Age vs Income")
plt.show()
```

```python
plt.figure(dpi=200)
sns.relplot(data=age_sal, x="Age", y="Annual Salary")
plt.title("Age vs Income")
plt.xlabel("Age (years)")
plt.ylabel("Income ($)")
plt.show()
```

```python
plt.figure(dpi=200,figsize=(8,8))
sns.relplot(data=age_sal, x="Age", y="Annual Salary")
plt.title("Age vs Income")
plt.xlabel("Age (years)")
plt.ylabel("Income ($)")
plt.show()
```
