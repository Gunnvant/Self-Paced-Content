## Modifying pandas plots

- Custom x axis and y axis labels and titles
- Changing the size of the plot

```python
### Pandas plots
import pandas as pd
weather = pd.read_csv("weather.csv")
weather['date'] = pd.to_datetime(weather['date'])
weather['months'] = weather['date'].dt.month
weather.head(2)
```

```python
atlanta = weather.query("city=='Atlanta'")[['TMIN','TMAX','months']]
atlanta['TMAX'].plot(kind='box')
```

```python
atlanta['TMAX'].plot(kind='box',figsize=(8,5), color='Red') ## changing the size and color
```

```python
atlanta['TMAX'].plot(kind='box',figsize=(8,5), title="Distribution of Maximum Temperatures") ## add title to the plot
```

```python
atlanta['TMAX'].plot(kind='box',figsize=(8,5), title="Distribution of Maximum Temperatures",grid=True) #add grids
```

```python
atlanta['TMAX'].plot(kind='box',figsize=(8,5), 
                     title="Distribution of Maximum Temperatures",
                     grid=True, ylabel="Max Temperature")
```

```python
age_sal = pd.read_csv("age_sal.csv")
```

```python
age_sal.plot(x="Age",y="Annual Salary",kind="scatter")
```

```python
age_sal.plot(x="Age",y="Annual Salary",kind="scatter",xlabel="Age (years)",ylabel="Salary ($)")
```

```python
age_sal.plot(x="Age",y="Annual Salary",kind="scatter",
             xlabel="Age (years)",ylabel="Salary ($)",
            title="Age vs Salary")
```

```python
age_sal.plot(x="Age",y="Annual Salary",kind="scatter",
             xlabel="Age (years)",ylabel="Salary ($)",
            title="Age vs Salary",figsize=(5,5))
```
