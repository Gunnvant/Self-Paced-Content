## Creating Bar Plots

import pandas as pd
data = pd.read_csv("movie_scores.csv")

```python
data.head()
```

```python
data.plot(x="MovieTitle",y="Tomatometer",kind="bar")
```

```python
data.plot(x="MovieTitle",y=["Tomatometer","AudienceScore"],kind="bar")
```

```python
import seaborn as sns
```

```python
sns.catplot(data,x="Tomatometer",y="MovieTitle",kind="bar")
```

```python
melted_data = pd.melt(data,id_vars='MovieTitle',value_vars =\
                      ['Tomatometer','AudienceScore'],value_name="score")
```

```python
melted_data
```

```python
sns.catplot(x='score', y='MovieTitle', hue='variable', data=melted_data, kind='bar')
```
