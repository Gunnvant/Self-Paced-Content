Read the data called `starbucks.csv`,

An ideal diet should contain optimum level of nutrients can you find out the names of the items on menu that contain:

- Upto 450 calories,
- Upto 40 g protein,
- Upto 10 g fat,
- Upto 40 g Carbs,
- Upto 5 g fibre,

Give the names of items on menu that satisfy the above criteria

```python
import pandas as pd
```

```python
data = pd.read_csv("starbucks.csv")
data.head(2)
```

```python
cond = '''Calories<=450 and Protein<=40 and Fat<=10 and Carb<=40 and Fiber<=5'''
data.query(cond)
```

```python
data.query(cond)['Name']
```

```python
## Can you rank the items based on the amount of protien?
data.query(cond).sort_values(by = "Protein",ascending=False)
```

```python
data.query(cond).sort_values(by = "Protein",ascending=False)['Name']
```
