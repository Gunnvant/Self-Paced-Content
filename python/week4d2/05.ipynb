Categorize sales based on its magnitude as given below:

- if sales> 5000, high
- if 2000<sales<=5000, medium
- if sales<2000, low


```python
def get_cat(val):
    if val > 5000:
        return "high"
    elif val<=5000 and val>2000:
        return "medium"
    else:
        return "low"
```

```python
cats = []
for val in data['Sales']:
    cats.append(get_cat(val))
```

```python
data['spend_cat'] = cats
```
