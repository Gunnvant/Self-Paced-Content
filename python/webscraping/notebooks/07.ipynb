```python
import requests
url = "https://gutendex.com/books/"
```

**API Docs [link](https://gutendex.com)**

```python
resp = requests.get(url).json()
```

```python
resp.keys()
```

```python
resp['results'][0:3]
```

Parse the `json` response to the table.

```python
ids = []
title = []
authors = []
subjects = []
langs = []
download_count = []
for row in resp['results']:
    ids.append(row['id'])
    title.append(row['title'])
    auths = [i['name'] for i in row['authors']]
    auths = ",".join(auths)
    authors.append(auths)
    subjects.append(",".join(row['subjects']))
    l = ",".join(row['languages'])
    langs.append(l)
    download_count.append(row['download_count'])
```

```python
import pandas as pd
pd.DataFrame({'ids':ids,'title':title,'authors':authors,'subjects':subjects,'langs':langs,'download_count':download_count})
```
