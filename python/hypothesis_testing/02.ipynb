Automate the process of doing the hypothesis test that we'd done for the city New Delhi for all the other cities present in the dataset

```python
import pandas as pd
import math
import scipy.stats
recent = pd.read_csv('recent_temp.csv')
global_data = pd.read_csv('global_temp.csv')
```

```python
def p_value(pop_mean,sample_mean,sample_stdev,sample_size):
    SE = sample_stdev/math.sqrt(sample_size)
    if pop_mean>sample_mean:
        pval = scipy.stats.norm(pop_mean,SE).cdf(sample_mean)
    else:
        pval = 1-scipy.stats.norm(pop_mean,SE).cdf(sample_mean)
    return pval
    
```

```python
results = []
for city in global_data['city'].unique():
    pop_mean = global_data[global_data['city']==city]['avg_temp'].values
    sample_mean = recent[recent['city']==city]['tavg'].mean()
    sample_stdev = recent[recent['city']==city]['tavg'].std()
    sample_size = recent[recent['city']==city]['tavg'].shape[0]
    pval = p_value(pop_mean,sample_mean,sample_stdev,sample_size)
    row = {'city':city,'p_val':pval,'pop_mean':pop_mean,'sample_mean':sample_mean}
    results.append(row)
```

```python
pd.DataFrame(results)
```
